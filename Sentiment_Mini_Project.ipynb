```python

# Sentiment Analysis Mini Project
# Using TF-IDF and Logistic Regression

from sklearn.feature_extraction.text import TfidfVectorizer 
#  TfidfVectorizer: converts text to numerical features
from sklearn.linear_model import LogisticRegression   
# LogisticRegression: simple machine learning model for classification
import pandas as pd
# pandas: to display the TF-IDF matrix nicely

# Small sample sentiment dataset
texts = [
    "I love this product",
    "This is amazing",
    "I am very happy",
    "I hate this",
    "This is terrible",
    "I am very sad"
]

labels = [1, 1, 1, 0, 0, 0]  # 1 = Positive, 0 = Negative

# Vectorization
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)



X_dense = X.toarray()  # Convert sparse → dense (small datasets only!)
df = pd.DataFrame(X_dense, columns=vectorizer.get_feature_names_out())

# Uncomment the line below to see TF-IDF features
# print(df)


# Train Logistic Regression model on the small dataset
model = LogisticRegression()
model.fit(X, labels)

# Function for prediction
def predict_sentiment(text):
    """
    Predict sentiment of input text.

    Parameters:
        text (str): Input sentence

    Returns:
        str: "Positive" or "Negative"
    """
    vec = vectorizer.transform([text])
    prediction = model.predict(vec)[0]
    return "Positive" if prediction == 1 else "Negative"

if __name__ == "__main__":
    # Test
    print("\n--- Sentiment Predictions ---\n")

print(predict_sentiment("I love this movie"))
print(predict_sentiment("I hate this film"))
print(predict_sentiment("This is wonderful"))
print(predict_sentiment("This is bad"))

print(predict_sentiment("The movie was fantastic and amazing!"))
# Expected: 'Positive'

print(predict_sentiment("I am disappointed and sad about this product."))
# Expected: 'Negative'

print("\nModel is ready for sentiment prediction.")
```
